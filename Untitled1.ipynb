```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

def wavelet_packet_decomposition(signal):
    # perform wavelet packet decomposition
    wp = pywt.WaveletPacket(signal, 'db4', mode='symmetric', maxlevel=6)
    coeffs = []
    for level in range(1, 6):
        level_coeffs = []
        # get the list of nodes at the current level
        nodes = wp.get_level(level, 'natural')
        # loop over the nodes and append their data to the level_coeffs list
        for node in nodes:
            level_coeffs.append(wp[node.path].data)
        coeffs.append(level_coeffs)
    return coeffs, wp

# generate a test signal
t = np.linspace(0, 1, 1024)
signal = np.sin(2*np.pi*10*t) + np.sin(2*np.pi*20*t) + np.sin(2*np.pi*50*t)

# perform wavelet packet decomposition and plot the results
coeffs, wp = wavelet_packet_decomposition(signal)

fig, axs = plt.subplots(len(coeffs), 1, figsize=(8, 12))
for i, level_coeffs in enumerate(coeffs):
    for j, node_coeffs in enumerate(level_coeffs):
        axs[i].plot(node_coeffs, label=f'Node {j+1}')
    axs[i].set_xlabel('Sample Index')
    axs[i].set_ylabel('Amplitude')
    axs[i].set_title(f'Level {i+1}')
    axs[i].legend()
plt.tight_layout()
plt.show()
```
